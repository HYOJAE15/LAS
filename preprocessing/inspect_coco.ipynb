```python
root_dir = 'Z:/05. Data/02. Training&Test/013. General Concrete Damage/06. COCO(Data used in CHALK)/crack/ver.0.1(overlap)'
subset = 'train'

instances_json_path = os.path.join(root_dir, 'annotations', f'instances_{subset}2018.json')
images_path = os.path.join(root_dir, f'{subset}2018')

coco_dataset = CocoDataset(instances_json_path, images_path)

coco_dataset.display_info()
coco_dataset.display_licenses()
coco_dataset.display_categories()
```

```python
# set original path 

delete_idx = np.zeros(len(coco_dataset.coco['images']))

for img_num, img in enumerate(coco_dataset.coco['images']):
    clear_output(wait=True)
    try:
        coco_dataset.display_image(img_num+1)
        input_ = str(input("Do you want to keep this image in the original path? [Y/N]"))

        if (input_ == 'Y') or (input_ == 'y') :            
            print('This image will be remained in the original path.')
            sleep(0.5)
        elif (input_ == 'N') or (input_ == 'n'):            
            print('This image and all of its annotations are permenantly deleted from the original path.')
            delete_idx[img_num] = 1
            print("Num of Images to be deleted :" + str(np.sum(delete_idx)))
            sleep(0.5)
            
    except KeyError :
        print("This image doesn't have an annotation file or an original image. It will be deleted from the list.")
        delete_idx[img_num] = 1
        print("Num of Images to be deleted :" + str(np.sum(delete_idx)))
        
    except FileNotFoundError:      
        print("This image doesn't have an annotation file or an original image. It will be deleted from the list.")
        delete_idx[img_num] = 1
        print("Num of Images to be deleted :" + str(np.sum(delete_idx)))
        
        
    
    
    
        
```
